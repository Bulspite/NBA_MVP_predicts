```python
import pandas as pd
```

```python
mvp=pd.read_csv("mvp_v1.csv")
mvp
```

```python
mvp=mvp[["Player","Year","Pts Won","Pts Max","Share"]]
mvp.head()
```

```python
player=pd.read_csv("player_stats.csv",index_col=0)
player
```

```python
player["Player"]=player["Player"].str.replace("*","",regex=False)
```

```python
player[25:27]
```

```python
def isolate(df):
    if df.shape[0]== 1:
        return df
    else:
        row=df[df["Tm"]=="TOT"]
        row["Tm"]=df.iloc[-1,:]["Tm"]
        return row

player=player.groupby(["Player","Year"]).apply(isolate)
```

```python
player.index=player.index.droplevel()
player.index=player.index.droplevel()
player
```

```python
comb_player = player.merge(mvp,how="outer",on=["Player", "Year"])
comb_player.head()
```

```python
comb_player[["Pts Won","Pts Max","Share"]] = comb_player[["Pts Won","Pts Max","Share"]].fillna(0)
comb_player
```

```python
teams=pd.read_csv("team_stats.csv",index_col=0)
teams.head(10)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
teams['Team'] = teams['Team'].str.replace("*","",regex=False)
```

```python
teams.head(10)
```

```python
nickname ={}
with open("nicknames.txt") as f:
    lines=f.readlines()
    for line in lines[1:]:
        abb,name =line.replace("\n","").split(",")
        nickname[abb] = name
```

```python
nickname
```

```python
comb_player["Team"]=comb_player["Tm"].map(nickname)
comb_player
```

```python
comb_player.drop("Rk",axis=1,inplace = True)
```

```python
comb_player
```

```python
stat= comb_player.merge(teams,how='outer',on=["Team","Year"])
stat
```

```python
stat = stat.apply(pd.to_numeric,errors="ignore")
```

```python
stat.dtypes
```

```python
stat['GB'].unique()
```

```python
stat['GB'].replace("—",0,inplace=True)
```

```python
stat.to_csv("Overall_stat.csv")
```
