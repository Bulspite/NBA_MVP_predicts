```python
def gap(comb):
    actual = comb.sort_values("Share",ascending = False).head()
    predicts = comb.sort_values("Prediction",ascending = False)
    error = []
    success = 0 
    search = 1
    for index, row in predicts.iterrows():
        if row["Player"] in actual["Player"].values:
            success+=1
            error.append(success/search)
        search+=1
    return sum(error)/len(error)
```

```python
gap(comb_pre)
```

#### Prediction of each year based on previous years

```python
years = list(range(1991,2022))
```

#### Function for each year predictions

```python
x= []
all_predicts = []
for y in years[5:]:
    train = stats[stats["Year"] < y]
    test = stats[stats["Year"] == y]
    reg.fit(train[feats],train["Share"])
    pre = reg.predict(test[feats])
    predf = pd.DataFrame(pre,columns= ["Prediction"], index = test.index)
    comb_pre = pd.concat([test[["Player","Share"]],predf], axis =1)
    all_predicts.append(comb_pre)
    x.append(gap(comb_pre))
```

```python
sum(x)/len(x)
```

#### Function for difference in each year Predicted VS Actual ranks

```python
def dif(comb_pre):
    comb_pre = comb_pre.sort_values("Share",ascending = False)
    comb_pre["Rank"] = list(range(1,comb_pre.shape[0]+1))
    comb_pre = comb_pre.sort_values("Prediction" ,ascending = False)
    comb_pre["Predicted Rank"] = list(range(1,comb_pre.shape[0]+1))
    comb_pre["Diff"] = comb_pre["Rank"] - comb_pre["Predicted Rank"]
    return comb_pre
```

```python
dif_rank = dif(all_predicts[1])
dif_rank[dif_rank["Rank"]<=5 ].sort_values(("Diff"), ascending = False)
```

#### Function to get previous year predictions from any point

```python
def backpredicts(stats,model,year,predicts):
    x= []
    all_predicts = []
    for y in years[5:]:
        train = stats[stats["Year"] < y]
        test = stats[stats["Year"] == y]
        model.fit(train[feats],train["Share"])
        pre = model.predict(test[feats])
        predf = pd.DataFrame(pre,columns= ["Prediction"], index = test.index)
        comb_pre = pd.concat([test[["Player","Share"]],predf], axis =1)
        comb_pre = dif(comb_pre)
        all_predicts.append(comb_pre)
        x.append(gap(comb_pre))
    return sum(x)/len(x), x, pd.concat(all_predicts)
```
